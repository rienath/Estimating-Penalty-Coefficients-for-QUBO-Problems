```python
import matplotlib.pyplot as plt
import numpy as np
import math

plt.style.use('ieee')
```

```python
%matplotlib inline
```

```python
x = np.linspace(-10, 10, 100)
y = 1/(1 + np.exp(-x))
  
plt.plot(x, y)
plt.xlabel("ΔE")
plt.ylabel("Probability")
plt.grid()
```

```python
x = np.linspace(-10, 10, 100)
y = (x >= 0)
  
plt.plot(x, y)
plt.xlabel("ΔE")
plt.ylabel("Probability")
plt.grid()
```

```python
x = np.linspace(-10, 10, 100)
y = x * 0 + 0.5
  
plt.plot(x, y)
plt.xlabel("ΔE")
plt.ylabel("Probability")
plt.grid()
```

```python
x = np.linspace(-10, 10, 100)
t = 10**-5
y0 = 1/(1 + np.exp(-x/t))
t = 1
y1 = 1/(1 + np.exp(-x/t))
t = 3
y2 = 1/(1 + np.exp(-x/t))
t = 10
y3 = 1/(1 + np.exp(-x/t))
t = 10**5
y4 = 1/(1 + np.exp(-x/t))

plt.plot(x, y0, label = 'T=10^-5', color='cyan')
plt.plot(x, y1, label = 'T=1')
plt.plot(x, y2, label = 'T=3')
plt.plot(x, y3, label = 'T=10')
plt.plot(x, y4, label = 'T=10^5', color='blue')
plt.xlabel("ΔE")
plt.ylabel("Probability")
plt.legend()
plt.grid()
```

```python
x = np.linspace(-10, 10, 100)
t = 10**-5
y0 = 1/(1 + np.exp(-x/t))
t = 1
y1 = 1/(1 + np.exp(-x/t))
t = 3
y2 = 1/(1 + np.exp(-x/t))
t = 10
y3 = 1/(1 + np.exp(-x/t))
t = 10**5
y4 = 1/(1 + np.exp(-x/t))

y0[y0 > 0.5] = 1
y1[y1 > 0.5] = 1
y2[y2 > 0.5] = 1
y3[y3 > 0.5] = 1
y4[y4 > 0.5] = 1


plt.plot(x, y0, label = 'T=10^-5', color='cyan')
plt.plot(x, y1, label = 'T=1')
plt.plot(x, y2, label = 'T=3')
plt.plot(x, y3, label = 'T=10')
plt.plot(x, y4, label = 'T=10^5', color='blue')
plt.xlabel("ΔE")
plt.ylabel("Probability")
plt.legend()
plt.grid()
```
